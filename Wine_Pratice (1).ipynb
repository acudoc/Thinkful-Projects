```python
plt.plot(['points'])
plt.show()
```

```python
plt.plot(df['points'], color='purple')
plt.plot(df['price'], color='green')
plt.ylim([-0.1, 5.1])
plt.ylabel('Wines')
plt.title('Price and Rating')
plt.show()
```

```python
plt.scatter(x=df['price'], y=df['points'])
plt.show()
```

```python
plt.scatter(
    x=df['price'],
    y=df['variety'],
    color='purple',
    marker='x', s=100
)
plt.scatter(
    x=df['points'],
    y=df['variety'],
    color='green',
    marker='x', s=100
)
plt.show()
```

```python
plt.scatter(
    x=df['price'],
    y=df['country'],
    color='purple',
    marker='x', s=10
)
plt.scatter(
    x=df['points'],
    y=df['country'],
    color='green',
    marker='x', s=10
)    
plt.show()
```

```python
plt.scatter(
    x=df['points'],
    y=df['price'],
    color='purple',
    marker='x', s=10
)
plt.show()
```

```python
plt.scatter(
    x=df['price'],
    y=df['points'],
    color='green',
    marker='x', s=10
)
plt.show()
```

```python
df['avg_points'] = df.points.mean()
df['avg_price'] = df.price.mean()
df['price_sq'] = df['price'] ** 2
```

```python
plt.scatter(
    x=df['points'],
    y=df['avg_price'],
    color='purple',
    marker='x', s=10
)
plt.show()
```

```python
plt.scatter(
    x=df['price'],
    y=df['avg_points'],
    color='purple',
    marker='x', s=10
)
plt.show()    
```

```python
plt.plot(df['points'], color='purple')
plt.ylim([70,110])
plt.ylabel('Value')
plt.title('Price of Wine')
plt.show()
```

```python
plt.plot(df['price'], color='green')
plt.xlim([0, 1000])
plt.ylim([0, 300])
plt.ylabel('Values')
plt.title('Wine Cost')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['price'], color='green')
plt.ylabel('Values')
plt.title('Cost')

plt.subplot(1, 2, 2)
plt.scatter(x = df['price'], y = df['price_sq'], color='green')
plt.ylabel('Squared Values')
plt.title('Squared Wine Cost')

plt.tight_layout()
plt.show()
```
